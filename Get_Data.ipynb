```python
def split_data(SOURCE_DIR, TRAINING_DIR, VALIDATION_DIR, TRAIN_SIZE):

  source = os.listdir(SOURCE_DIR)

  shuffled = random.sample(source, len(source))

  training_number = int(len(shuffled) * TRAIN_SIZE)

  i = 0
  target = TRAINING_DIR

  for item in shuffled:
    item_source = os.path.join(SOURCE_DIR, item)
    if os.path.getsize(item_source) == 0:
      print(f'{item} is zero length, so ignoring.')
    else:
      copyfile(item_source, os.path.join(target, item))
      i += 1

    if i == training_number:
      target = VALIDATION_DIR

AEROBIC_STEPPERS_SOURCE_DIR = "./drive/MyDrive/gym_data2/aerobic_steppers/"
BARBELL_SOURCE_DIR = "./drive/MyDrive/gym_data2/barbell/"
BENCH_PRESS_SOURCE_DIR = "./drive/MyDrive/gym_data2/bench_press/"
CABLE_MACHINE_SOURCE_DIR = "./drive/MyDrive/gym_data2/cable_machine/"
DUMB_BELL_SOURCE_DIR = "./drive/MyDrive/gym_data2/dumb_bell/"
ELLIPTICAL_SOURCE_DIR = "./drive/MyDrive/gym_data2/elliptical/"
LEG_PRESS_MACHINE_SOURCE_DIR = "./drive/MyDrive/gym_data2/leg_press_machine/"
ROWING_MACHINE_SOURCE_DIR = "./drive/MyDrive/gym_data2/rowing_machine/"
SMITH_MACHINE_SOURCE_DIR = "./drive/MyDrive/gym_data2/smith_machine/"
STATIONARY_BIKE_SOURCE_DIR = "./drive/MyDrive/gym_data2/stationary_bike/"
TREADMILL_SOURCE_DIR = "./drive/MyDrive/gym_data2/treadmill/"

TRAINING_DIR = "./drive/MyDrive/gym_data2/training"
VALIDATION_DIR = "./drive/MyDrive/gym_data2/validation"

TRAINING_AEROBIC_STEPPERS_DIR = os.path.join(TRAINING_DIR, "aerobic_steppers/")
TRAINING_BARBELL_DIR = os.path.join(TRAINING_DIR, "barbell/")
TRAINING_BENCH_PRESS_DIR = os.path.join(TRAINING_DIR, "bench_press/")
TRAINING_CABLE_MACHINE_DIR = os.path.join(TRAINING_DIR, "cable_machine/")
TRAINING_DUMB_BELL_DIR = os.path.join(TRAINING_DIR, "dumb_bell/")
TRAINING_ELLIPTICAL_DIR = os.path.join(TRAINING_DIR, "elliptical/")
TRAINING_LEG_PRESS_MACHINE_DIR = os.path.join(TRAINING_DIR, "leg_press_machine/")
TRAINING_ROWING_MACHINE_DIR = os.path.join(TRAINING_DIR, "rowing_machine/")
TRAINING_SMITH_MACHINE_DIR = os.path.join(TRAINING_DIR, "smith_machine/")
TRAINING_STATIONARY_BIKE_DIR = os.path.join(TRAINING_DIR, "stationary_bike/")
TRAINING_TREADMILL_DIR = os.path.join(TRAINING_DIR, "treadmill/")

VALIDATION_AEROBIC_STEPPERS_DIR = os.path.join(VALIDATION_DIR, "aerobic_steppers/")
VALIDATION_BARBELL_DIR = os.path.join(VALIDATION_DIR, "barbell/")
VALIDATION_BENCH_PRESS_DIR = os.path.join(VALIDATION_DIR, "bench_press/")
VALIDATION_CABLE_MACHINE_DIR = os.path.join(VALIDATION_DIR, "cable_machine/")
VALIDATION_DUMB_BELL_DIR = os.path.join(VALIDATION_DIR, "dumb_bell/")
VALIDATION_ELLIPTICAL_DIR = os.path.join(VALIDATION_DIR, "elliptical/")
VALIDATION_LEG_PRESS_MACHINE_DIR = os.path.join(VALIDATION_DIR, "leg_press_machine/")
VALIDATION_ROWING_MACHINE_DIR = os.path.join(VALIDATION_DIR, "rowing_machine/")
VALIDATION_SMITH_MACHINE_DIR = os.path.join(VALIDATION_DIR, "smith_machine/")
VALIDATION_STATIONARY_BIKE_DIR = os.path.join(VALIDATION_DIR, "stationary_bike/")
VALIDATION_TREADMILL_DIR = os.path.join(VALIDATION_DIR, "treadmill/")

if len(os.listdir(TRAINING_AEROBIC_STEPPERS_DIR)) > 0:
  for file in os.scandir(TRAINING_AEROBIC_STEPPERS_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_BARBELL_DIR)) > 0:
  for file in os.scandir(TRAINING_BARBELL_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_BENCH_PRESS_DIR)) > 0:
  for file in os.scandir(TRAINING_BENCH_PRESS_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_CABLE_MACHINE_DIR)) > 0:
  for file in os.scandir(TRAINING_CABLE_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_DUMB_BELL_DIR)) > 0:
  for file in os.scandir(TRAINING_DUMB_BELL_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_ELLIPTICAL_DIR)) > 0:
  for file in os.scandir(TRAINING_ELLIPTICAL_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_LEG_PRESS_MACHINE_DIR)) > 0:
  for file in os.scandir(TRAINING_LEG_PRESS_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_ROWING_MACHINE_DIR)) > 0:
  for file in os.scandir(TRAINING_ROWING_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_SMITH_MACHINE_DIR)) > 0:
  for file in os.scandir(TRAINING_SMITH_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_STATIONARY_BIKE_DIR)) > 0:
  for file in os.scandir(TRAINING_STATIONARY_BIKE_DIR):
    os.remove(file.path)
if len(os.listdir(TRAINING_TREADMILL_DIR)) > 0:
  for file in os.scandir(TRAINING_TREADMILL_DIR):
    os.remove(file.path)

if len(os.listdir(VALIDATION_AEROBIC_STEPPERS_DIR)) > 0:
  for file in os.scandir(VALIDATION_AEROBIC_STEPPERS_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_BARBELL_DIR)) > 0:
  for file in os.scandir(VALIDATION_BARBELL_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_BENCH_PRESS_DIR)) > 0:
  for file in os.scandir(VALIDATION_BENCH_PRESS_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_CABLE_MACHINE_DIR)) > 0:
  for file in os.scandir(VALIDATION_CABLE_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_DUMB_BELL_DIR)) > 0:
  for file in os.scandir(VALIDATION_DUMB_BELL_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_ELLIPTICAL_DIR)) > 0:
  for file in os.scandir(VALIDATION_ELLIPTICAL_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_LEG_PRESS_MACHINE_DIR)) > 0:
  for file in os.scandir(VALIDATION_LEG_PRESS_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_ROWING_MACHINE_DIR)) > 0:
  for file in os.scandir(VALIDATION_ROWING_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_SMITH_MACHINE_DIR)) > 0:
  for file in os.scandir(VALIDATION_SMITH_MACHINE_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_STATIONARY_BIKE_DIR)) > 0:
  for file in os.scandir(VALIDATION_STATIONARY_BIKE_DIR):
    os.remove(file.path)
if len(os.listdir(VALIDATION_TREADMILL_DIR)) > 0:
  for file in os.scandir(VALIDATION_TREADMILL_DIR):
    os.remove(file.path)

training_size = .7

split_data(AEROBIC_STEPPERS_SOURCE_DIR, TRAINING_AEROBIC_STEPPERS_DIR, VALIDATION_AEROBIC_STEPPERS_DIR, training_size)
split_data(BARBELL_SOURCE_DIR, TRAINING_BARBELL_DIR, VALIDATION_BARBELL_DIR, training_size)
split_data(BENCH_PRESS_SOURCE_DIR, TRAINING_BENCH_PRESS_DIR, VALIDATION_BENCH_PRESS_DIR, training_size)
split_data(CABLE_MACHINE_SOURCE_DIR, TRAINING_CABLE_MACHINE_DIR, VALIDATION_CABLE_MACHINE_DIR, training_size)
split_data(DUMB_BELL_SOURCE_DIR, TRAINING_DUMB_BELL_DIR, VALIDATION_DUMB_BELL_DIR, training_size)
split_data(ELLIPTICAL_SOURCE_DIR, TRAINING_ELLIPTICAL_DIR, VALIDATION_ELLIPTICAL_DIR, training_size)
split_data(LEG_PRESS_MACHINE_SOURCE_DIR, TRAINING_LEG_PRESS_MACHINE_DIR, VALIDATION_LEG_PRESS_MACHINE_DIR,  training_size)
split_data(ROWING_MACHINE_SOURCE_DIR, TRAINING_ROWING_MACHINE_DIR, VALIDATION_ROWING_MACHINE_DIR, training_size)
split_data(SMITH_MACHINE_SOURCE_DIR, TRAINING_SMITH_MACHINE_DIR, VALIDATION_SMITH_MACHINE_DIR, training_size)
split_data(STATIONARY_BIKE_SOURCE_DIR, TRAINING_STATIONARY_BIKE_DIR, VALIDATION_STATIONARY_BIKE_DIR,  training_size)
split_data(TREADMILL_SOURCE_DIR, TRAINING_TREADMILL_DIR, VALIDATION_TREADMILL_DIR, training_size)
```
